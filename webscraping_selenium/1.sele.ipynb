```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.sanmar.com/NEW/c/new"
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(10)
html_content = driver.page_source
driver.quit()
soup = BeautifulSoup(html_content, "html.parser")
products = soup.find_all("div", class_="product-item product px-3 product-new")

product_data = []

for product in products:

  product_data.append({
      "Product Name": product.find("div", class_="product-name").a.text.strip() if product.find("div", class_="product-name") else None,
      "Product Code": product.find("input", class_="product-code")["value"].strip() if product.find("input", class_="product-code") else None,
      "Image URL": product.find("div", class_="product-image").find("img")["src"].strip('/') if product.find("div", class_="product-image").find("img") else None,
      "Style Number": product.find("span", class_="style-number").text.strip() if product.find("span", class_="style-number") else None,
      "Display Price": product.find("span", class_="displayPrice").text.strip() if product.find("span", class_="displayPrice") else None
  })

product_data
```
